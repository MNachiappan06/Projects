```python
#String columns that are required to be converted to numeric for calculation purposes
cols_tonumeric=['Pos', 'Seasons','Points', 'GamesPlayed', 'GamesWon',
       'GamesDrawn', 'GamesLost', 'GoalsFor', 'GoalsAgainst', 'Champion',
       'Runner-up', 'Third', 'Fourth', 'Fifth', 'Sixth', 'T','BestPosition']
```

```python
#Converting the above mentioned columns to numeric
F1[cols_tonumeric]=F1[cols_tonumeric].apply(pd.to_numeric)
```

```python
#Write a function with name “Goal_diff_count” which should return all the
#teams with their Goal Differences. Using the same function, find the team which has maximum and 
#minimum goal difference. (5 points)
def Goal_diff_count():
    F1['Goal_diff'] = F1['GoalsFor'] - F1.GoalsAgainst
    print(F1.Goal_diff)
    print("Team with Max goal diff:",F1.Goal_diff.max() )
    print("Team with Max goal diff:",F1.Goal_diff.min() )
```
